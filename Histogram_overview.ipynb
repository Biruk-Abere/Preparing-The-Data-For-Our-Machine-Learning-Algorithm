```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ages = [18 ,19 ,21, 25 , 26 , 26 , 30 , 32 ,38 ,45 ,56]
plt.hist(ages  , bins = 5)
```

```python
# if we want to see the bins well 
plt.hist(ages  , bins = 5 , edgecolor = 'black')
```

```python
# Another method is 
ages = [18 ,19 ,21, 25 , 26 , 26 , 30 , 32 ,38 ,45 ,56]
bins = [10 , 20 , 30 , 40 , 50  , 60]
plt.hist(ages , bins = bins , edgecolor = 'black')
```

```python
# if we want to exclude some of the ages 
ages = [18 ,19 ,21, 25 , 26 , 26 , 30 , 32 ,38 ,45 ,56]
bins = [ 30 , 40 , 50  , 60]
plt.hist(ages , bins = bins , edgecolor = 'black')
```
